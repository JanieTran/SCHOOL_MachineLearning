Next, **write a function** to plot multiple histograms of all 14 columns in the dataset. Hint: you will need to loop through the column indices. Besides, you will need to set up a grid first where all of the histograms will be plotted. Do that by using the following command:

`f, axes = plt.subplots(4, 4, figsize = (20, 15), sharex = False)`

```python
def plot_df_hist(df, ncols):
    # Calculate number of plot rows from number of features and number of plot columns specified
    nrows = len(df.columns) // ncols + 1
    
    # Set up plot grid list
    f, axes = plt.subplots(nrows, ncols, figsize = (20, 15), sharex = False)
    row = 0
    plot = 0
    
    # Iterate through columns of dataframe
    for col in df.columns:
        sns.distplot(df[col], hist = True, color = 'r', kde = True, ax = axes[row][plot])
        plot += 1
        if plot == ncols:
            plot = 0
            row += 1
```

```python
plot_df_hist(bostonHouseFrame, 4)
```

**Can you explain all the parameters of this function? What do they mean? Can you draw the histograms to look like this and explain all parameters that are used to draw these histograms:**

> *ANSWER: The function takes in 2 parameters: `df` is the dataframe needed to be plotted, and `ncols` to specify the number of columns to display the resulted plots.*

**Now, write a similar function to plot a series of box-plots.**

```python
def plot_df_box(df, ncols):
    # Calculate number of plot rows from number of features and number of plot columns specified
    nrows = len(df.columns) // ncols + 1
    
    # Set up plot grid list
    f, axes = plt.subplots(nrows, ncols, figsize = (20, 15), sharex = False)
    row = 0
    plot = 0
    
    # Iterate through columns of dataframe
    for col in df.columns:
        sns.boxplot(df[col], color = 'g', ax = axes[row][plot])
        plot += 1
        if plot == ncols:
            plot = 0
            row += 1
```

```python
plot_df_box(bostonHouseFrame, 4)
```
