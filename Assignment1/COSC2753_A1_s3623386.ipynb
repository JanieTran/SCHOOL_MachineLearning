## Baseline

```python
from sklearn.model_selection import train_test_split

X = train.drop(['TARGET_deathRate', 'id'], axis = 1)
Y = train['TARGET_deathRate']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 191)

print('x_train:', x_train.shape)
print('y_train:', y_train.shape)
print('x_test:', x_test.shape)
print('y_test:', y_test.shape)
```

Result dataframe for comparison

```python
results_comparison = pd.DataFrame()
```

Helper function for model fitting

```python
from sklearn.metrics import mean_absolute_error

def fit_model(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train)
    preds = model.predict(x_test)
    return mean_absolute_error(preds, y_test)
```

Chosen algorithms are **Linear Regression, Support Vector Machines (SVM),** and **Nearest Neighbours**

```python
from sklearn.linear_model import LinearRegression
from sklearn import svm
from sklearn.neighbors import KNeighborsRegressor

linear_regression = LinearRegression()
svr = svm.SVR(gamma = 'scale')
knn = KNeighborsRegressor(n_neighbors = 10)

models = {
    'linear': linear_regression,
    'svr': svr,
    'knn': knn,
}
```

Helper function to fit all models

```python
def run_models(x_train, x_test, y_train, y_test, result_col):
    # Initialise result column in results dataframe
    results_comparison[result_col] = 0
    
    # Run each model
    for key, model in models.items():
        mae = fit_model(model, x_train, x_test, y_train, y_test)
        results_comparison.loc[key, result_col] = mae
        print(key, 'MAE:', mae)
```

```python
run_models(x_train, x_test, y_train, y_test, 'baseline')
```

```python
results_comparison.plot(kind = 'bar')
```

## Model Improvement

### Cross Validation

```python
from sklearn.model_selection import KFold

kfold = KFold(n_splits = 5, random_state = 191, shuffle = True)

# Function to run K-Fold cross validation
def run_cv(X, Y, result_col):
    fold = 0
    for train_index, test_index in kfold.split(X):
        x_train, x_test = X.loc[train_index], X.loc[test_index]
        y_train, y_test = Y.loc[train_index], Y.loc[test_index]

        fold += 1
        print('------Fold', fold)

        col_name = result_col + '_' + str(fold)
        run_models(x_train, x_test, y_train, y_test, col_name)
        
# Get average results from 5 folds
def get_avg_results(prefix):
    regex = prefix + '_.*'
    results_comparison[prefix] = results_comparison.filter(regex = regex).mean(axis = 1)
    results_comparison.drop(results_comparison.filter(regex = regex).columns, axis = 1, inplace = True)
    results_comparison.plot(kind = 'bar')
```

```python
run_cv(X, Y, 'cv')
```

```python
results_comparison.plot(kind = 'bar')
```

```python
get_avg_results('cv')
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
X_scaled = pd.DataFrame(scaler.transform(X), columns = X.columns)

X_scaled.head()
```

```python
run_cv(X_scaled, Y, 'cv_scale')
```

```python
get_avg_results('cv_scale')
```

### Reduce Skewness

```python
right_skewed = ['avgAnnCount', 'avgDeathsPerYear', 'incidenceRate', 'medIncome', 'popEst2015', 'povertyPercent',
                'studyPerCap', 'PctNoHS18_24', 'PctBachDeg18_24', 'PctBachDeg25_Over', 'PctUnemployed16_Over',
                'PctPublicCoverageAlone', 'PctBlack', 'PctAsian', 'PctOtherRace', 'BirthRate']

left_skewed = ['AvgHouseholdSize', 'PercentMarried', 'PctHS25_Over', 'PctEmployed16_Over',
               'PctPrivateCoverage', 'PctWhite']

X_reduce_skew = X.copy()
```

```python
X_reduce_skew[right_skewed] = np.log1p(X_reduce_skew[right_skewed])
X_reduce_skew[left_skewed] = X_reduce_skew[left_skewed] ** 2
```

```python
plt.figure()
X_reduce_skew.hist(figsize = (15, 15))
plt.show()
```

```python
run_cv(X_reduce_skew, Y, 'cv_skew')
```

```python
get_avg_results('cv_skew')
```

```python
results_comparison
```

### Scale + Reduce Skewness

```python
# De-skew
X_scale_norm = X.copy()
X_scale_norm[right_skewed] = np.log1p(X_scale_norm[right_skewed])
X_scale_norm[left_skewed] = X_scale_norm[left_skewed] ** 2

# Scale
scaler.fit(X_scale_norm)
X_scale_norm = pd.DataFrame(scaler.transform(X_scale_norm), columns = X_scale_norm.columns)

run_cv(X_scale_norm, Y, 'cv_scale_skew')
```

```python
get_avg_results('cv_scale_skew')
```
