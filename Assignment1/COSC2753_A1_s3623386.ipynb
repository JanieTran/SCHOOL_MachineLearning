## Helper Functions

Function to plot multiple histograms

```python
def plot_df_hist(df, ncols, w_fig = 20, h_fig = 25, pad = 4, w_pad = 0.5, h_pad = 1.0):
    # Calculate number of plot rows from number of features and number of plot columns specified
    nrows = len(df.columns) // ncols + 1
    
    # Set up plot grid list
    f, axes = plt.subplots(nrows, ncols, figsize = (w_fig, h_fig), sharex = False)
    plt.tight_layout(pad = 0.4, w_pad = w_pad, h_pad = h_pad)
    row = 0
    plot = 0
    
    # Iterate through columns of dataframe
    for col in df.columns:
        sns.distplot(df[col], hist = True, color = 'r', kde = True, ax = axes[row][plot])
        plot += 1
        if plot == ncols:
            plot = 0
            row += 1
```
